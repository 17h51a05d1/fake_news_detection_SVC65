```python
import matplotlib.pyplot as plt
import numpy as np

x1 = [24000, 50000, 65618]
x2 = [24000, 50000, 65618]

y1 = [99, 94, 95]
y2 = [99, 95, 97]

plt.xlim([25000, 70000])
plt.ylim([94, 100])

font1 = {'family':'serif','color':'black','size':15}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x1, y1, color='r', label="model 1")
plt.plot(x2, y2, color='g', label="model 2")

plt.xlabel('Size of dataset taken', fontdict = font2)
plt.ylabel('Accuracy', fontdict = font2)
plt.title('Accuracy difference between model 1 and model 2', fontdict = font1)

plt.legend(["model 1", "model 2"])

plt.grid()
plt.show()
```
