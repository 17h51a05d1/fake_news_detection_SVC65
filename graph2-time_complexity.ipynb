```python
import numpy as np
import matplotlib.pyplot as plt

x1 = [24000, 50000, 65618]
x2 = [24000, 50000, 65618]

y1 = [45, 225, 850]
y2 = [32, 190, 480]

plt.xlim([25000, 70000])

font1 = {'family':'serif','color':'black','size':15}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x1, y1, color='r', label="model 1")
plt.plot(x2, y2, color='g', label="model 2")

plt.xlabel('Size of dataset taken', fontdict = font2)
plt.ylabel('Time Complexity', fontdict = font2)
plt.title('Time complexity difference between model 1 and model 2', fontdict = font1)

plt.legend(["model 1", "model 2"])

plt.grid()
plt.show()
```
